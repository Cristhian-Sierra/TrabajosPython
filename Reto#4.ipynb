Camilo ha decidido dejar atrás su viejo computador y comprar uno nuevo por partes para poder él mismo armarlo. Además, también quiere darle un nuevo aspecto a su puesto de juego y trabajo cambiando el escritorio y la silla. Para esto, Camilo crea una lista de los elementos que le hacen falta para su nuevo setup electrónico. Al hacer diferentes cotizaciones por tiendas virtuales (No todos los elementos se encuentran en una solo tienda), Camilo se dio cuenta que cada tienda responde con un archivo JSON que contiene la información de los productos disponibles junto con su precio.

Para ayudar a Camilo, usted deberá hacer un programa que dado un JSON que contiene la información de una cotización de una tienda virtual y la lista de elementos que le hacen falta a Camilo por comprar, de como resultado el precio total de la compra y los elementos que puede comprar en esa tienda.

Entrada

En la primera línea, un texto que corresponde al JSON de los elementos disponibles a la venta con su precio. En la segunda línea, la lista de los elementos, separados por un espacio, que le hacen falta a Camilo por comprar.

Salida

En la primera línea, el precio total de los productos que se pueden comprar en esa tienda. En la segunda línea, la lista de productos que se pueden comprar en esa tienda.

```python
import json


def recorrer_lista(datosT,productos):
  listaF=[]
  listaT=[]
  str1=productos.split(" ")
  listaP=list(str1)
  data=json.loads(datosT)
  listaT=[]


  for j  in listaP:
    for key in data:
      if key ==j:
        listaF.append(j)
        listaT.append(data[key])
  sumaT=sum(listaT)
  print("\n",sumaT)            
  return listaF



productosT=input("Digite productos en tienda: ")  
product=input("Digite productos deseados: ")
strlistaP=" ".join(recorrer_lista(productosT,product))
print(strlistaP)
```
