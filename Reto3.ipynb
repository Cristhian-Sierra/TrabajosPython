Camilo y un nuevo juego

Camilo y Juan se aburrieron del juego que Juan propuso inicialmente, por lo que Camilo propone un nuevo juego (algo aburrido). En este caso, Camilo propone que en una bolsa van a estar escritas en papeles diferentes las vocales minúsculas (a, e, i, o, u) y los símbolos aritméticos (+, -, *, /). Una vez la bolsa tenga todos los papeles, tanto Camilo y Juan van a apostar cual letra o símbolo va a salir más veces seguido de la bolsa.

Para empezar el juego, alguno de los dos va a sacar un papel y va a escribir el símbolo que tiene ese papel en una hoja, luego volverá a insertar el papel en la bolsa y repetirá el proceso varias veces. Al final, se tendrá una lista de símbolos y letras que se obtuvieron al sacar los papeles de la bolsa. Luego ellos van a contar el símbolo que más veces seguido esté en la lista antes de encontrar otro símbolo y ganará aquel que haya acertado el símbolo o letra que esté más veces seguido.

Usted deberá crear un programa que realice el conteo de los símbolos seguidos que se encuentran en la lista creada después de sacar los papeles de la bolsa.

Entrada

Una lista de símbolos y letras separadas por un espacio, que simula la lista de símbolos que se obtuvo después de sacarlos de la bolsa.

Salida

En la primera línea, una lista de caracteres que muestra los símbolos y las letras que salieron de la bolsa. En la segunda línea, una lista de números que refleja la cantidad de veces que cada letra o símbolo salió seguida de la bolsa.

```python
from itertools import groupby


def contador(listaF):
  grupos=groupby(listaF)
  listaC=[]
  for k,g in grupos:
    listaC.append(len(list(g)))
  return listaC
  

def elemento(listaF):
  grupos=groupby(listaF)
  listaE=[]
  for k,g in grupos:
    listaE.append(str (k))
  return listaE


listaS=input("Digite los caracteres sacados de la bolsa ")
listaS2=listaS.split(" ")
lie=elemento(listaS2)
lic=contador(listaS2)
le=" ".join(lie)
lc=" ".join(map(str, lic))
print(le)
print(lc)
```
