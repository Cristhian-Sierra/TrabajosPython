¿Quién ganará?

En una tarde de juegos, Camilo y Juan se aburren de los juegos convencionales de mesa que tienen. Por esto, Juan propone un nuevo juego de azar con base en algunas letras del abecedario y símbolos aritméticos. Juan le dice a Camilo que en una bolsa van a estar escritas en papeles diferentes las consonantes minúsculas (b, c, d, f, g, h, j, k, l, m, n, p, q, r, s, t, u, v, w, x, y, z) y los símbolos aritméticos (+, -, *, /, %). Una vez la bolsa tenga las letras y los símbolos Juan propone el siguiente juego:

·         Camilo debe sacar cualquier cantidad de papeles de la bolsa y escribir en un papel las letras o símbolos que le salieron, luego debe introducir de nuevo los papeles en la bolsa.

·         Juan realiza exactamente el mismo procedimiento anterior.

·         Una vez que cada uno tenga su propio conjunto de letras y símbolos, en cada turno del juego se sacará un papel de la bolsa y si la letra o el símbolo que se obtiene hace parte del conjunto de letras y símbolos de un jugador, a ese jugador se le suma un punto y luego se vuelve a introducir el papel en la bolsa.

Como Juan es un poco tramposo, Camilo le propone una forma de llevar las cuentas del juego a medida que se van sacando papeles de la bolsa: Sí con la letra o símbolo que se obtiene de la bolsa pasa a ganar Camilo en un papel se escribe la letra C, si por el contrario pasa a ganar Juan en el papel se escribe la letra J, pero si con el nuevo papel se llega a un empate se imprime la letra T. Al final del juego, se espera tener tantas C, J o T como papeles se saquen de la bolsa.


Entrada

Usted deberá crear un programa que replique el juego planteado por Juan. Como entrada recibirá tres datos: El primero es el conjunto de caracteres (letras y símbolos) que tiene Juan, el segundo dato es el conjunto de caracteres de Camilo y el último dato son los caracteres que se sacaron de la bolsa turno por turno en orden.

Salida

El programa deberá retornar la cadena de caracteres con las letras C, J o T que representen como evolucionó el juego a medida que se iban sacando papeles de la bolsa.

```python

#bolsa = ["b", "c", "d", "f", "g", "h", "j", "k", "l", "m", "n", "p", "q", "r", "s", "t", "v", "w", "x", "y", "z", "+" , "-", "*","/", "%"]  #se define una lista

def puntaje(caJ,caC,listaT):
  listaF=""
  acumJ=0
  acumC=0
  for i in range(len(listaT)):
    if (caJ.count(listaT[i])>0):
      acumJ+=1
    if (caC.count(listaT[i])>0):
      acumC+=1
      
    if (acumJ>acumC):
      listaF+='J'
    elif (acumC>acumJ):
        listaF+='C'
    elif (acumJ==acumC):
      listaF+='T'

  return listaF
  



casJ=input()
casC=input()
listaT=input()


punt=puntaje(casJ,casC,listaT).replace("'","")
print(punt)
    

```
