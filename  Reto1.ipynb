Existe una ciudad muy organizada en donde la construcción de edificios se hace de a tres, es decir, siempre que se construya un edificio se deben construir dos más. Además de esto, la ciudad está dividida en cuatro zonas en donde se pueden ubicar los edificios. La elección de la zona dependerá de la altura de los tres edificios.

Una compañía desea construir conjuntos de tres edificios en la ciudad y contrata a un arquitecto que debe diseñar cada trio de edificios con ciertas condiciones con el fin de ahorrar costos. La constructora le dice al arquitecto que sólo conocerá la altura del primer edificio. Además, si se le quitan 4 metros a la altura del segundo edificio se tendrá la altura del primero duplicada. Por último, la suma de alturas entre el primero y el segundo equivaldrá a cinco veces la altura del tercero.

La zona en la cual deberá quedar el conjunto de edificios dependerá de la altura del tercer edificio y está organizada así:

·         Zona 'uno': Para cuando el tercer edificio está entre 0 y 20 metros

·         Zona 'dos': Para cuando el tercer edificio está entre 21 y 30 metros

·         Zona 'tres': Para cuando el tercer edificio está entre 31 y 50 metros

·         Zona 'cuatro': Para cuando el tercer edificio tiene más de 50 metros

 

Con esta información, la constructora lo contrata a usted para apoyar al arquitecto y le pide que haga un programa que encuentre en que zona de la ciudad se tiene que construir un conjunto de tres edificios si se conoce la altura del primer edificio. El programa tendrá que imprimir en la primera línea la altura de los tres edificios en orden (primer, segundo, tercer edificio) y en la segunda línea se tiene que mostrar la zona en la cual se construirán los edificios:

Entrada

La entrada es un número entero que representa la altura del primer edificio.

Salida

Tres enteros separados por espacio que representan las alturas de los edificios. En la siguiente línea en letras la zona en donde se construirán los edificios.

Ejemplo
Entrada: 18
Salida: 18 40 11
 Uno






```python
def altura_edificio2(e1):
  return (2*e1) +4

def altura_edificio3(e1,e2):
  return (e1+e2)//5

def zona_construir(e3):
  if (e3 >=0 and e3 <=20):
    return "uno"

  elif (e3 >20 and e3 <=30):
    return "dos"

  elif (e3 >30 and e3 <=50):
    return "tres"

  else: 
    return "cuatro"

edificio1=int(input("Digite altura de primer edificio (entera) "))
edificio2=altura_edificio2(edificio1)
edificio3=altura_edificio3(edificio1,edificio2)
zonaC=zona_construir(edificio3)

print(edificio1," ",edificio2," ", edificio3)
print(zonaC)








  
```
