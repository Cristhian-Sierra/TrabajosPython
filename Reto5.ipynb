Camilo y Juan arman un nuevo PC

Al igual que Camilo, Juan ha decidido armar un nuevo PC. Para esto se pone de acuerdo con Camilo de intercambiar algunas partes de PC que ellos ya no usen y que necesiten. Para esto plantean una lista de partes como la siguiente:

[‘disco_HDD’, ‘silla’, ‘lector_CD’, ‘juegos’, ‘caja’, ‘teclado’, ‘memorias’, ‘windows’, ‘pantallas’]

Para más facilidad a la hora del intercambio, Camilo y Juan crean grupos de partes y les asignan un nombre. Por ejemplo, el grupo ‘perifericos’ contiene partes como teclado, pantallas, parlantes, etc. El grupo ‘software’ contiene partes como juegos, Windows, etc. El grupo ‘dispositivos’ contiene partes como caja, disco_HDD, memorias, etc. Para la lista de partes mostrada anteriormente la lista de grupos es la siguiente:

[‘dispositivos’, ‘accesorios’, ‘accesorios’, ‘software’, ‘dispositivos’, ‘perifericos’, ‘dispositivos’, ‘software’, ‘perifericos’]

Sólo se puede cambiar partes que pertenezcan al mismo grupo.

Para ayudar a Camilo y Juan, se le pide a usted hacer un módulo en Python con nombre nuevopc.py que va a ser usado por otros programadores y que debe tener como mínimo las siguientes funciones:

1.       La función “grupos” que dada una lista en donde se muestra los grupos de cada parte (cada posición i de la lista es el grupo de la parte i), genera una lista con los grupos únicos existentes. Por ejemplo, si se recibe una lista como la siguiente:

[‘dispositivos’, ‘accesorios’, ‘accesorios’, ‘software’, ‘dispositivos’, ‘perifericos’, ‘dispositivos’, ‘software’, ‘perifericos’] -> Lista de grupos

Devuelve:

[‘dispositivos’, accesorios’, software’, ’perifericos’] -> Lista de grupos únicos

 

2.    La función “necesito_del_grupo” que dada una lista con los índices de la lista de las partes, la lista de los grupos de cada parte y el nombre de un grupo, retorne una lista con las partes que se necesiten de la categoría dada. Por ejemplo, si se reciben los datos siguientes:

[0,1,3,6,8] -> Son los índices de cada parte en la lista de partes

[‘dispositivos’, ‘accesorios’, ‘accesorios’, ‘software’, ‘dispositivos’, ‘perifericos’, ‘dispositivos’, ‘software’, ‘perifericos’] -> Lista de grupos

‘dispositivos’ -> El grupo que interesa

Retorna:

[0,6] -> Se necesitan las partes que están en la posición 0 y 6 de la lista de partes que pertenecen al grupo ‘dispositivos’

 

3.    La función “sirven_a_juan” que dada una lista con los índices de la lista de partes que le sobran a Camilo y una lista con los índices de partes que ya tiene Juan, devolver la lista de índices de las partes que le sirven a Juan de Camilo. Por ejemplo, si se reciben las siguientes listas:

[0,1,5,6,8] -> A Camilo le sobran las partes con índice 0, 2, 4 y 6.

[0,2,5,6,7] -> Juan ya tiene las partes con índice 1, 2, 5, 6

Devolver:

[1,8] -> De las partes que tiene Camilo sólo le sirven a Juan las que están en los índices 1 y 8

 

4.    Al final, Camilo y Juan se ponen de acuerdo de que los grupos ya no les interesan, por lo que ahora se pueden cambiar partes de diferentes grupos. Además, cada uno de los dos tiene una lista con las partes que puede cambiar y aquellas partes que no estén en esa lista son los que necesita. Para esto, el módulo que usted está construyendo deberá tener una nueva función llamada “cuantas_cambian” que recibirá dos listas. La primera serán los índices de las partes que puede cambiar Camilo y la segunda los índices de las partes que puede cambiar Juan. La función deberá retornar la cantidad de partes que se pueden cambiar. Por ejemplo, si se reciben las siguientes dos listas:

[0,1,2,3,6,7] -> Camilo puede cambiar las partes con índice 0, 1, 2, 3, 6 y 7.

[2,3,5,6,8] -> Juan puede cambiar las partes con índice 2, 3, 5 y 6.

Retorna:

2 -> Sólo se pueden cambiar dos partes, ya que a Juan le interesan las partes con índice 0, 1 y 7 de Camilo, pero a Camilo sólo le interesan las partes con índice 5 y 8 porque los otros ya los tiene. Por ende, sólo se pueden hacer dos cambios.

 

Observaciones:

```python
#Modulo nuevos

def grupos(lista):
  listaF=[]
  for i in lista:
    if i not in (listaF):
      listaF.append(i)
  return listaF

def necesito_del_grupo(listaI,listaG,grupo):
  listaFI=[]
  for i in  listaI:
    for j in range(len(listaG)): 
      if i == j and grupo==listaG[j]:
        listaFI.append(j)
  return listaFI

def sirven_a_juan(listaICa,listaIJu):
  listaT=[]
  for i in listaICa:
      if i not in listaIJu:
        listaT.append(i)
  return listaT

def cuantas_cambian(listaC,listaJ):
  cont=0
  if len(listaC)<len(listaJ):
    for i in listaC:
        if i  not in listaJ:
          cont+=1
    return cont
  else:
    for j in listaJ:
        if j  not in listaC:
          cont+=1
    return cont

```
